```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import os

# This line ensures plots are displayed directly in the notebook
%matplotlib inline

class ExpenseTracker:
    def __init__(self, filename='expenses.csv'):
        self.filename = filename
        # Data storage using CSV 
        if not os.path.exists(self.filename):
            df = pd.DataFrame(columns=['Date', 'Category', 'Amount', 'Description'])
            df.to_csv(self.filename, index=False)

    def add_expense(self):
        """Record daily expenses including date, category, amount, and description """
        print("\n--- Log New Expense ---")
        date = input("Enter date (YYYY-MM-DD) or press Enter for today: ")
        if not date:
            date = datetime.now().strftime("%Y-%m-%d")
        
        # Categorize spending into predefined groups 
        print("Categories: Food, Travel, Bills, Shopping, Others") 
        category = input("Enter Category: ").capitalize()
        
        try:
            amount = float(input("Enter Amount: "))
        except ValueError:
            print("Invalid amount. Please enter a number.")
            return

        description = input("Enter Description: ")

        # Store data in CSV format 
        new_data = pd.DataFrame([[date, category, amount, description]])
        new_data.to_csv(self.filename, mode='a', header=False, index=False)
        print(f" Expense of {amount} added to {category}!")

    def generate_summary(self):
        """Generate monthly summaries and insights [cite: 120, 124]"""
        df = pd.read_csv(self.filename)
        if df.empty:
            print("No data available to analyze.")
            return

        df['Date'] = pd.to_datetime(df['Date'])
        df['Month'] = df['Date'].dt.to_period('M')

        # Filter for the current month 
        current_month = datetime.now().strftime("%Y-%m")
        monthly_df = df[df['Month'] == current_month]

        if monthly_df.empty:
            print(f"No expenses found for the month: {current_month}")
            return

        total_spent = monthly_df['Amount'].sum()
        print(f"\n--- Summary for {current_month} ---")
        print(f"Total Monthly Spending: {total_spent}")

        # Detect the highest spending category 
        category_totals = monthly_df.groupby('Category')['Amount'].sum()
        highest_cat = category_totals.idxmax()
        highest_val = category_totals.max()
        
        print(f" Highest Spending Category: {highest_cat} ({highest_val})")
        
        # Suggest areas to reduce spending based on patterns 
        print(f" Suggestion: Consider reviewing your {highest_cat} habits to save money!")

        self.visualize_expenses(category_totals)

    def visualize_expenses(self, category_totals):
        """pie chart is employed here to derscribe category wise representation"""
        plt.figure(figsize=(8, 6))
        category_totals.plot(kind='pie', autopct='%1.1f%%', startangle=140)
        plt.title('Monthly Expense Distribution by Category')
        plt.ylabel('')  
        plt.show()

# Execution loop
tracker = ExpenseTracker()
while True:
    print("\n===== SMART EXPENSE TRACKER =====")
    print("1. Add Expense")
    print("2. View Monthly Summary & Insights")
    print("3. Exit")
    choice = input("Select an option (1-3): ")

    if choice == '1':
        tracker.add_expense()
    elif choice == '2':
        tracker.generate_summary()
    elif choice == '3':
        print("Exiting tracker...")
        break
    else:
        print("Invalid choice.")
```
